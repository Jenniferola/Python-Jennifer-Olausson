```python
import csv
import math
import matplotlib.pyplot as plt 

# Changed the file extension of datapoints.txt to .csv

pokemons = []
testpoints_x_values_width = []
testpoints_y_values_height = []
pichu_datapoints = []
pichu_x_values_width = []
pichu_y_values_height =[]
pikachu_datapoints = []
pikachu_x_values_width = []
pikachu_y_values_height =[]

testpoint_1_PIKACHU_distances = []
testpoint_1_sorted_PIKACHU_distances = []
testpoint_1_PICHU_distances = []
testpoint_1_sorted_PICHU_distances = []
testpoint_2_PIKACHU_distances = []
testpoint_2_sorted_PIKACHU_distances = []
testpoint_2_PICHU_distances = []
testpoint_2_sorted_PICHU_distances = []
testpoint_3_PIKACHU_distances = []
testpoint_3_sorted_PIKACHU_distances = []
testpoint_3_PICHU_distances = []
testpoint_3_sorted_PICHU_distances = []
testpoint_4_PIKACHU_distances = []
testpoint_4_sorted_PIKACHU_distances = []
testpoint_4_PICHU_distances = []
testpoint_4_sorted_PICHU_distances = []

user_input_PIKACHU_distances = []
user_input_sorted_PIKACHU_distances = []
user_input_PICHU_distances = []
user_input_sorted_PICHU_distances = []
user_inputs = []

def save_data_from_file():
    #reads datapoints from cvs-file and adds to a list with 3 key-value pairs. 

    with open("..\\data\\datapoints.csv") as file:
        reader = csv.DictReader(file)
    
        for row in reader:
            pokemons.append({"width" : float(row["(width (cm)"].strip()), "height" : float(row[" height (cm)"].strip()), "label" : int(row[" label (0-pichu"].strip())})


def plot_matplotlib(pichu_x, pichu_y, pikachu_x, pikachu_y):
    #plots datapoints and 4 testpoints into matplotlib

    testpoints_x_values_width = [25,24.2,22,20.5]
    testpoints_y_values_height = [32,31.5,34,34]

    plt.title("Pichu and Pikachu")
    plt.xlabel("Width (cm)")
    plt.ylabel("Height(cm)")

    plt.scatter(pichu_x, pichu_y, color = "orange", label = "Pichu") 
    plt.scatter(pikachu_x, pikachu_y, color = "green", label = "Pikachu")
    plt.scatter(testpoints_x_values_width, testpoints_y_values_height, color ="pink", label ="Testpoints")
    plt.legend()


def calculate_euclidean_distance(x1,y1,x2,y2): 
    #calculates the euclidean distance between a testpoint and datapoints.
    global eucladian_distance
    eucladian_distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return eucladian_distance


def user_inputs(text):
    #user input for testpoint. Wrong message if not a number or a number below 0. 
    while True:
        try:
            user_input = (float(input(text)))

            if user_input < 0:
                print("wrong, please write a positive number")
                continue
            else:
                return user_input

        except ValueError:
            print("Wrong, please write an integear or float")
        

save_data_from_file()


# the datapoints of pikachu and pichu are being devided into separate lists and also by width (x) and height (y).
pichu_datapoints = [item for item in pokemons if item["label"] == 0]
pikachu_datapoints = [item for item in pokemons if item["label"] == 1]


for row in pichu_datapoints:
    pichu_x_values_width.append(row["width"])

for row in pichu_datapoints:
    pichu_y_values_height.append(row["height"])

for row in pikachu_datapoints:
    pikachu_x_values_width.append(row["width"])

for row in pikachu_datapoints:
    pikachu_y_values_height.append(row["height"])


plot_matplotlib(pichu_x_values_width, pichu_y_values_height, pikachu_x_values_width,pikachu_y_values_height)



testpoints_ordered = [25,32,24.2,31.5,22,34,20.5,34]

# calculats if the 4 datapoints are classified as pikachu or pichu:
# testpoint 1
for i in range(75):
    calculate_euclidean_distance(testpoints_ordered[0],testpoints_ordered[1],pikachu_x_values_width[i],pikachu_y_values_height[i])
    testpoint_1_PIKACHU_distances.append(eucladian_distance)
    testpoint_1_sorted_PIKACHU_distances = sorted(testpoint_1_PIKACHU_distances)

    calculate_euclidean_distance(testpoints_ordered[0], testpoints_ordered[1],pichu_x_values_width[i],pichu_y_values_height[i])
    testpoint_1_PICHU_distances.append(eucladian_distance)
    testpoint_1_sorted_PICHU_distances = sorted(testpoint_1_PICHU_distances)

if testpoint_1_sorted_PIKACHU_distances[0] < testpoint_1_sorted_PICHU_distances[0]: 
    result_testpoint_1 = "Testpoint 1 is classified as Pikachu."
else:   
    result_testpoint_1 = "Testpoint 1 is classified as Pichu"


# testpoint 2
for i in range(75):
    calculate_euclidean_distance(testpoints_ordered[2],testpoints_ordered[3],pikachu_x_values_width[i],pikachu_y_values_height[i])
    testpoint_2_PIKACHU_distances.append(eucladian_distance)
    testpoint_2_sorted_PIKACHU_distances = sorted(testpoint_2_PIKACHU_distances)

    calculate_euclidean_distance(testpoints_ordered[2], testpoints_ordered[3],pichu_x_values_width[i],pichu_y_values_height[i])
    testpoint_2_PICHU_distances.append(eucladian_distance)
    testpoint_2_sorted_PICHU_distances = sorted(testpoint_2_PICHU_distances)

if testpoint_2_sorted_PIKACHU_distances[0] < testpoint_2_sorted_PICHU_distances[0]: 
    result_testpoint_2 = "Testpoint 2 is classified as Pikachu." 
else:   
    result_testpoint_2 = "Testpoint 2 is classified as Pichu"


# testpoint 3
for i in range(75):
    calculate_euclidean_distance(testpoints_ordered[4],testpoints_ordered[5],pikachu_x_values_width[i],pikachu_y_values_height[i])
    testpoint_3_PIKACHU_distances.append(eucladian_distance)
    testpoint_3_sorted_PIKACHU_distances = sorted(testpoint_3_PIKACHU_distances)

    calculate_euclidean_distance(testpoints_ordered[4], testpoints_ordered[5],pichu_x_values_width[i],pichu_y_values_height[i])
    testpoint_3_PICHU_distances.append(eucladian_distance)
    testpoint_3_sorted_PICHU_distances = sorted(testpoint_3_PICHU_distances)

if testpoint_3_sorted_PIKACHU_distances[0] < testpoint_3_sorted_PICHU_distances[0]: 
    result_testpoint_3 = "Testpoint 3 is classified as Pikachu." 
else:   
    result_testpoint_3 = "Testpoint 3 is classified as Pichu"


# testpoint 4
for i in range(75):
    calculate_euclidean_distance(testpoints_ordered[6],testpoints_ordered[7],pikachu_x_values_width[i],pikachu_y_values_height[i])
    testpoint_4_PIKACHU_distances.append(eucladian_distance)
    testpoint_4_sorted_PIKACHU_distances = sorted(testpoint_4_PIKACHU_distances)

    calculate_euclidean_distance(testpoints_ordered[6], testpoints_ordered[7],pichu_x_values_width[i],pichu_y_values_height[i])
    testpoint_4_PICHU_distances.append(eucladian_distance)
    testpoint_4_sorted_PICHU_distances = sorted(testpoint_4_PICHU_distances)

if testpoint_4_sorted_PIKACHU_distances[0] < testpoint_4_sorted_PICHU_distances[0]: 
    result_testpoint_4 = "Testpoint 4 is classified as Pikachu."
else:   
    result_testpoint_4 = "Testpoint 4 is classified as Pichu."

print(f"{result_testpoint_1}\n{result_testpoint_2}\n{result_testpoint_3}\n{result_testpoint_4}\n")


# exercise 1) classifie a testpoint from the user
user_testpoint_x = user_inputs("Exercise 1) Write a testpoint (x value):")
user_testpoint_y = user_inputs("Exercise 1) Write a testpoint (y value):")


for i in range(75):
    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y, pikachu_x_values_width[i],pikachu_y_values_height[i])
    user_input_PIKACHU_distances.append(eucladian_distance)
    user_input_sorted_PIKACHU_distances = sorted(user_input_PIKACHU_distances)

    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y,pichu_x_values_width[i],pichu_y_values_height[i])
    user_input_PICHU_distances.append(eucladian_distance)
    user_input_sorted_PICHU_distances = sorted(user_input_PICHU_distances)

if user_input_sorted_PIKACHU_distances[0] < user_input_sorted_PICHU_distances[0]: 
    result_user_input = "your testpoint is classified as pikachu."
else:   
    result_user_input = "your testpoint is classified as pichu."


print(f"exercise 1) {result_user_input}")


# exercise 2) classifie a testpoint from the user by a majority vote of the 10 closest testpoints. 
user_input_PIKACHU_distances.clear()
user_input_sorted_PIKACHU_distances.clear()
user_input_PICHU_distances.clear()
user_input_sorted_PICHU_distances.clear()

user_testpoint_x = user_inputs("Exercise 2) Write a testpoint (x value):")
user_testpoint_y = user_inputs("Exercise 2) Write a testpoint (y value):")

for i in range(75):
    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y, pikachu_x_values_width[i],pikachu_y_values_height[i])
    user_input_PIKACHU_distances.append(eucladian_distance)
    user_input_sorted_PIKACHU_distances = sorted(user_input_PIKACHU_distances)

    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y,pichu_x_values_width[i],pichu_y_values_height[i])
    user_input_PICHU_distances.append(eucladian_distance)
    user_input_sorted_PICHU_distances = sorted(user_input_PICHU_distances)

for i in range(75):
    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y, pikachu_x_values_width[i],pikachu_y_values_height[i])
    user_input_PIKACHU_distances.append(eucladian_distance)
    user_input_sorted_PIKACHU_distances = sorted(user_input_PIKACHU_distances)

    calculate_euclidean_distance(user_testpoint_x, user_testpoint_y,pichu_x_values_width[i],pichu_y_values_height[i])
    user_input_PICHU_distances.append(eucladian_distance)
    user_input_sorted_PICHU_distances = sorted(user_input_PICHU_distances)


added_2_lists = user_input_sorted_PIKACHU_distances + user_input_sorted_PICHU_distances
sorted_list = sorted(added_2_lists)

ten_shortest_distances = []
count=0
for number in sorted_list:
    if count <10:
        ten_shortest_distances.append(sorted_list[count])
        count += 1
    else: 
        break

closest_to_pikachu = []
closest_to_pichu = []

count=0
for number in ten_shortest_distances:
    if number in user_input_PIKACHU_distances:
        closest_to_pikachu.append(number)
    else:
        closest_to_pichu.append(number)
    count +=1


if len(closest_to_pikachu) > len(closest_to_pichu):
    majority_result= (f"your testpoint is classified as pikachu (itÂ´s a majority of {len(closest_to_pikachu)}")
else:
    majority_result =(f"your testpoint is classified as pichu (itÂ´s a majority of {len(closest_to_pichu)}")


print(f"exercise 2) {majority_result} votes).")

```

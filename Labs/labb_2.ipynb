```python
import csv
import matplotlib.pyplot as plt

pokemons = []

pichu_datapoints = []
pichu_x_values_width = []
pichu_y_values_height =[]

pikachu_datapoints = []
pikachu_x_values_width = []
pikachu_y_values_height =[]

# First I changed the file-extension of datapoints.txt to .csv instead.


# Open data and add to a list with 3 dictionaries.
with open("..\\data\\datapoints.csv") as file:
    reader = csv.DictReader(file)

    # Adding the data from datapoints.csv to a list with 3 dictionaries. Strip() takes away the whitespeca and the strings are converted to float or int.
    for row in reader:
        pokemons.append({"width" : float(row["(width (cm)"].strip()), "height" : float(row[" height (cm)"].strip()), "label" : int(row[" label (0-pichu"].strip())})



# The values of pichu and pikachu seperated to two new lists. 
pichu_datapoints = [item for item in pokemons if item["label"] == 0]

pikachu_datapoints = [item for item in pokemons if item["label"] == 1]



# Made 2 new lists for x_values (width) and y_values (height) for pichu.
for row in pichu_datapoints:
    pichu_x_values_width.append(row["width"])

for row in pichu_datapoints:
    pichu_y_values_height.append(row["height"])



# Made 2 new lists for x_values (width) and y_values (height) for pikachu.
for row in pikachu_datapoints:
    pikachu_x_values_width.append(row["width"])

for row in pikachu_datapoints:
    pikachu_y_values_height.append(row["height"])





#Matplotlib layout


plt.title("Datapoints Pichu and Pikachu")
plt.xlabel("Width (cm)")
plt.ylabel("Height(cm)")

plt.scatter(pikachu_x_values_width, pikachu_y_values_height, color = "green", label = "Pikachu")
plt.scatter(pichu_x_values_width, pichu_y_values_height, color = "orange", label = "Pichu") 
plt.legend()





```

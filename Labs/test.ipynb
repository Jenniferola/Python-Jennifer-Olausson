```python

list1 = [1,2,3,4,5,6,]
list2 = [7,10,22,33,50,70]

print(f"Här är lista 1:\n{list1}")
print(f"Här är lista 2:\n{list2}")


print("De två listorna läggs ihop till en ny lista osorterade:")

list3 = list1 + list2

print(list3)

sorted_list = sorted(list3)

print(f"Listan sorteras: (det finns {len(sorted_list)} tal ursprungligen.)")

print(sorted_list)

ten_list = []

print("De 10 lägsta talen läggs till i en ny lista: ")

count=0
for number in sorted_list:
    if count <10:
        ten_list.append(sorted_list[count])
        count += 1
    else: 
        break

print(ten_list)

if_number_in_list1 = []
if_number_in_list2 = []




count=0
for number in ten_list:
    if number in list1:
        if_number_in_list1.append(number)
    else:
        if_number_in_list2.append(number)
    count +=1

print("Talen från lista1 som finns med i tio-listan radas upp här:")

print(if_number_in_list1)

print("Talen från lista2 som finns med i tio-listan radas upp här:")

print(if_number_in_list2)

if len(if_number_in_list1) > len(if_number_in_list2):
    print(f"List1 are a majority: {len(if_number_in_list1)}")
else:
    print(f"List2 are a majority: {len(if_number_in_list2)}")





```
